```python
import pandas as pd
from scipy.stats import anderson, kstest, cramervonmises, shapiro

# Define functions for each test

def anderson_darling_test(data):
    result = anderson(data)
    print(f"Anderson-Darling Test Statistic: {result.statistic}")
    print(f"Anderson-Darling Critical Values: {result.critical_values}")
    print(f"Anderson-Darling Significance Levels: {result.significance_level}\n")

def kolmogorov_smirnov_test(data):
    statistic, p_value = kstest(data, 'norm')
    print(f"Kolmogorov-Smirnov Test Statistic: {statistic}")
    print(f"Kolmogorov-Smirnov P-Value: {p_value}\n")

def cramervonmises_test(data):
    result = cramervonmises(data, 'norm')
    print(f"Cramer-von Mises Test Statistic: {result.statistic}")
    print(f"Cramer-von Mises P-Value: {result.pvalue}\n")

def shapiro_wilk_test(data):
    statistic, p_value = shapiro(data)
    print(f"Shapiro-Wilk Test Statistic: {statistic}")
    print(f"Shapiro-Wilk P-Value: {p_value}\n")

# Load data and extract the 'Pre-molt' column
pre_molt_sizes_lab =df[df['Location'] == 'Lab']['Pre-molt']

# Run all tests
anderson_darling_test(pre_molt_sizes_lab)
kolmogorov_smirnov_test(pre_molt_sizes_lab)
cramervonmises_test(pre_molt_sizes_lab)
shapiro_wilk_test(pre_molt_sizes_lab)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

post_field = df[df['Location'] == 'Field']['Post-molt']
pre_field = df[df['Location'] == 'Field']['Pre-molt']

# Perform Linear Least Squares for treatment data
slope, intercept, r_value, p_value, std_err = linregress(post_field, pre_field)

# Calculate predicted values using the linear model
predicted_pre = slope * post_field + intercept

# Calculate residuals
residuals_f = pre_field-predicted_pre

# Create a figure for the plots
plt.figure(figsize=(12, 6))

# Scatterplot with regression line
plt.subplot(1, 2, 1)
plt.scatter(post_field, pre_field, color='blue', alpha=0.7, label='Data')
plt.plot(post_field, predicted_pre, color='red', label='Linear Regression')

# Add labels and title
plt.title('Linear Least Squares: pre-molt vs post-molt for field data')
plt.xlabel('post-molt')
plt.ylabel('pre-molt')
plt.legend()

# Residual plot
plt.subplot(1, 2, 2)
plt.scatter(predicted_pre, residuals_f, color='green', alpha=0.7)

# Add a horizontal line for zero residuals
plt.axhline(0, color='black', linestyle='--', label='Zero Residuals Line')

# Add labels and title
plt.title('Residuals Plot')
plt.xlabel('Predicted pre molt')
plt.ylabel('Residuals')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()

# Print Pearson's R-squared value
r_squared = r_value**2
print(f"Pearson's R-squared value: {r_squared}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
import statsmodels.api as sm

# Utilizing cross-validation to estimate prediction accuracy of Linear models
field_post_molt_sizes =df[df['Location'] == 'Field']['Post-molt']
field_pre_molt_sizes =df[df['Location'] == 'Field']['Pre-molt']
lab_post_molt_sizes =df[df['Location'] == 'Lab']['Post-molt']
lab_pre_molt_sizes =df[df['Location'] == 'Lab']['Pre-molt']

# Combine the data for Lab and Field crabs
field_post_molt_data = np.array(field_post_molt_sizes)
lab_post_molt_data = np.array(lab_post_molt_sizes)

all_post_molt_sizes = np.concatenate([field_post_molt_data, lab_post_molt_data])
all_pre_molt_sizes = np.concatenate([field_pre_molt_sizes, lab_pre_molt_sizes])

# Create a DataFrame for combined data
all_locations = ['Field'] * len(field_post_molt_sizes) + ['Lab'] * len(lab_post_molt_sizes)
combined_data = pd.DataFrame({
    'Location': all_locations,
    'Post-molt': all_post_molt_sizes,
    'Pre-molt': all_pre_molt_sizes
})

# Perform 5-fold cross-validation for both Lab and Field crabs
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize arrays to store predicted values and true labels
predicted_values = np.array([])
true_labels = np.array([])

# Loop through each fold
for train_idx, test_idx in kfold.split(combined_data):

    # Split data into training and testing sets for each fold
    train_data = combined_data.iloc[train_idx]
    test_data = combined_data.iloc[test_idx]

    # Fit the model on training data (use sm.OLS here)
    model = sm.OLS(train_data['Pre-molt'], sm.add_constant(train_data['Post-molt']))
    results = model.fit()

    # Predict on the testing data
    predictions = results.predict(sm.add_constant(test_data['Post-molt']))

    # Store the predicted values and true labels
    predicted_values = np.concatenate((predicted_values, predictions))
    true_labels = np.concatenate((true_labels, test_data['Pre-molt'].values))

# Calculate R-squared manually
r_squared = r2_score(true_labels, predicted_values)

print("Cross-validated R-squared:", r_squared)
```
